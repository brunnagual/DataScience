# Operadores Aritméticos


```python
# EXERCÍCIO 1
# Declaração de variáveis - 'nome_da_variavel = valor' - atribuição
# int(*******) força a variável a ser do tipo int
# função input('mensagem do input') - abre uma caixa para entrada de dados do usuário com a mensagem no lugar dos ********



n1 = int(input('Entre com o número 1: '))
n2 = int(input('Entre com o número 2: '))

print('')
print('A soma dos numeros (n1+n2) é: ', n1+n2)
print('A subtração dos numeros (n1-n2) é: ', n1-n2)
print('A multiplicação dos numeros (n1*n2) é: ', n1*n2)
print('A divisão de (n1/n2) é: ', n1/n2)
print('n1 elevado a n2 (n1**n2) é: ', n1**n2)
print('o resto da divisão de n1 por n2 (n1%n2) é: ', n1%n2)
```

```python
#EXERCÍCIO 2

n1 = int(input('Entre com o número que deseja saber a tabuada: '))

# Estrutura do FORMAT
# Dentro dos {} tem um termo {X:YK}
# X  é o primeiro termo dentro do format - format(X,Z,.....)
# Z é o segundo termo (lembrando que em programação os termos não iniciam em 1, mas sim em 0) - ex({0:2d} se refere ao n1 dentro do termo format(n1,n1*1))
# Y é o número de casas disponíveis para exibição do valor
# K é o tipo da variável - se for D é um inteiro (sem casas decimais), se for F é um float e os senhores (as) definem a quantidade de casas decimais após a vírgula

print('')
print('{0:2d} *  1 = {1:3d}'.format(n1,n1*1))
print('{0:2d} *  2 = {1:3d}'.format(n1,n1*2))
print('{0:2d} *  3 = {1:3d}'.format(n1,n1*3))
print('{0:2d} *  4 = {1:3d}'.format(n1,n1*4))
print('{0:2d} *  5 = {1:3d}'.format(n1,n1*5))
print('{0:2d} *  6 = {1:3d}'.format(n1,n1*6))
print('{0:2d} *  7 = {1:3d}'.format(n1,n1*7))
print('{0:2d} *  8 = {1:3d}'.format(n1,n1*8))
print('{0:2d} *  9 = {1:3d}'.format(n1,n1*9))
print('{0:2d} * 10 = {1:3d}'.format(n1,n1*10))
```

```python
# DESAFIO!

# não tem muita diferença para os anteriores
# vale lembrar que no comparativo o CASE SENSITIVE (diferenciação de maiúsculo e minúsulo) faz diferença, então se não colocar o operador com letra maiúscula vai dar OPERAÇÃO INVÁLIDA
# os operadore aritméticos podem ser lançados diretamente dentro do format, sem necessidade de criação de variáveis auxiliares
# a escolha do FLOAT ao invés do INT foi feita devido a não saber o que o usuário vai digitar, pois é uma calculadora


var1 = float(input('Entre com o primeiro valor: '))
var2 = float(input('Entre com o segundo valor: '))
operacao = input('entre com a operação: ')

if operacao == 'Soma':
  print('var1 + var2 = {}'.format(var1+var2))
elif operacao == 'Subtracao':
  print('var1 - var2 = {}'.format(var1-var2))
elif operacao == 'Multiplicacao':
  print('var1 * var2 = {}'.format(var1*var2))
elif operacao == 'Divisao':
  print('var1 / var2 = {}'.format(var1/var2))
elif operacao == 'Potencia':
  print('var1 ** var2 = {:.5f}'.format(var1**var2))
else:
  print('Operação digitada é inválida')
```

```python
#Exercício 3
#Faça um programa que pede para o usuário inserir um produto e salvar este produto numa lista.
#Caso o produto já esteja na lista informe uma mensagem de erro e não cadastre o produto duplicado.

lista_produtos = []

aux = 0
n = True

while n:

  if len(lista_produtos) == 0:
    produto = input('Entre com o produto que deseja cadastrar: ')
    lista_produtos.append(produto)
    
    continuar = input('Deseja continuar inserindo produtos?')
    if  continuar == 's' or continuar == 'S':
      n = True
    else:
      n = False 
    print('')
  else:
    produto = input('Entre com o produto que deseja cadastrar: ')
  
    for i in lista_produtos:
      if produto == i:
        aux = -1
        break

    if aux == -1:
      print('Produto já cadastrado\n')
    else:
      lista_produtos.append(produto)
    
    aux = 0
    
    continuar = input('Deseja continuar inserindo produtos?')
    if  continuar == 's' or continuar == 'S':
      n = True
    else:
      n = False 
    print('')

print('\n', lista_produtos)
```

```python
#Exercício 1
#Crie uma lista de dicionários contendo as notas de todos os alunos de uma sala (5 alunos). No dicionário é para ter os campos:
#nome:
#nota1:
#nota2:
#media:
#status:
#Peça para o usuário digitar as notas de cada aluno
#Tire a média das notas e adicione no dicionário
#Se média for maior ou igual a 7, coloque "aprovado" no status, caso contrário, coloque "reprovado"
#Calcule a média das médias e imprima para o professor.



alunos = [{'aluno': 'Aluno 1', 'nota1' : 0, 'nota2' : 0, 'media' : 0, 'Status' : False }, 
          {'aluno': 'Aluno 2', 'nota1' : 0, 'nota2' : 0, 'media' : 0, 'Status' : False }, 
          {'aluno': 'Aluno 3', 'nota1' : 0, 'nota2' : 0, 'media' : 0, 'Status' : False }, 
          {'aluno': 'Aluno 4', 'nota1' : 0, 'nota2' : 0, 'media' : 0, 'Status' : False }, 
          {'aluno': 'Aluno 5', 'nota1' : 0, 'nota2' : 0, 'media' : 0, 'Status' : False }, 
          ]

somaMedias = 0


for i, a in enumerate(alunos):
  a['nota1'] = float(input(f"Entre com a nota 1 do aluno {a['aluno']}: "))
  a['nota2'] = float(input(f"Entre com a nota 2 do aluno {a['aluno']}: "))
  a['media'] = (a['nota1'] + a['nota2'])/2
  somaMedias+=a['media']

  if a['media'] >= 7:
    a['Status'] = True


for a in alunos:
  print(a['aluno'],f"\nNota 1: {a['nota1']}", f"\nNota 2: {a['nota2']}", f"\nMédia: {a['media']:.2f}")
  if a['Status'] == True:
    print('Aluno Aprovado')
    print()
  else:
    print('Aluno Reprovado')
    print()

    
print(f'A media dos alunos foi: {(somaMedias/5):.2f}')
print()



```
