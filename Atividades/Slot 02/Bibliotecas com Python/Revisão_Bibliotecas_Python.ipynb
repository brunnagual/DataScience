```python
times = [
  'Flamengo',
  'Corinthians',
  'Brasiliense'
]

times
```

```python
def cria_partidas(times):
  partida = []
  for time_casa in times:
    for time_visitante in times:
      if time_casa != time_visitante:
        partida.append((time_casa, time_visitante))
  return partida


cria_partidas(times)
```

```python
def filtra_partida(partida, time):
  for p in partida:
    if p[0] == time or p[1] == time:
      print(p)
```

```python
filtra_partida(partida,'Flamengo')
```

```python
filtra_partida(partida,'Corinthians')
```

```python
filtra_partida(partida,'Brasiliense')
```
