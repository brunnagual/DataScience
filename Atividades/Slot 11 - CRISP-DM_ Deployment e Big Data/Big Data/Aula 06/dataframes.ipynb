```python
from pyspark.sql import SparkSession
```

```python
# Instanciar a sessão spark e depois coloca a função builder,mostrando para sessão que ela precisa ser criada ou se não existe que ele precisa criar com a função getOrcreate.

spark = SparkSession.builder.appName("Meu primeiro Dataframe").getOrCreate()
```

```python
# Ler o arquivo e transformando em um dataframe.
df = spark.read.csv('./base_de_dados.csv')
```

```python
# referenciando df e mostrando através da função show (mostra o conteudo)
df.show()
```

```python
# Outra forma de ler um CSV para tentar mostrar pelo header onde ele entende que a primeira linha da minhas colunas são é o nome das colunas.
df = spark.read.csv(
    './base_de_dados.csv',
    header=True
)
```

```python
df.show()
```

```python
# Utilizando a função sep onde passamos qual caractere separa as colunas.
df = spark.read.csv(
    './base_de_dados.csv',
    header=True,
    sep=';'
)
```

```python
df.show()
```

```python
# Print todas as colunas e os tipos delas.
df.printSchema()
```
