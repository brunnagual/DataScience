# Exercicio Prático Clustering Hierarquico

## Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.neighbors import NearestNeighbors
from sklearn.datasets import load_iris
```

## Dataset

```python
df, df['target'] = pd.DataFrame(load_iris().data,columns= load_iris().feature_names), load_iris().target
```

```python
df
```

```python
df.columns
```

```python
FEATURES = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)',
       'petal width (cm)']
```

```python
agg = AgglomerativeClustering(affinity = 'euclidean', linkage = 'ward')
```

```python
df['cluster_hierarquico'] = agg.fit_predict(df[FEATURES])
```

```python
df.groupby('target')['cluster_hierarquico'].value_counts()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16,8))
axes[0].scatter(df['sepal length (cm)'], df['petal length (cm)'], c=df['target'])
axes[1].scatter(df['sepal length (cm)'], df['petal length (cm)'], c=df['cluster_hierarquico'], cmap=plt.cm.Set1)
axes[0].set_title('Target', fontsize=18)
axes[1].set_title('Hierarchical', fontsize=18)
```

```python
plt.figure(figsize = (20,12))
dendrogram = sch.dendrogram(sch.linkage(df[FEATURES], method  = "ward"))
plt.title('Dendrogram')
plt.xlabel('Flowers')
plt.ylabel('Euclidean distances')
plt.show()
```

### KMeans

```python
from sklearn.cluster import KMeans
```

```python
# fittando vários modelos e armazenando a inércia em uma lista
SSE = []
for cluster in range(1,20):
    kmeans = KMeans(n_clusters = cluster, init='k-means++')
    kmeans.fit(df[FEATURES])
    SSE.append(kmeans.inertia_)
# convertendo os resultados em um dataframe e plotando
frame = pd.DataFrame({'Cluster':range(1,20), 'SSE':SSE})
plt.figure(figsize=(12,6))
plt.plot(frame['Cluster'], frame['SSE'], marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
```

```python
km = KMeans(3)
```

```python
df['cluster_kmeans'] = km.fit_predict(df[FEATURES])
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16,8))
axes[0].scatter(df['sepal length (cm)'], df['petal length (cm)'], c=df['target'])
axes[1].scatter(df['sepal length (cm)'], df['petal length (cm)'], c=df['cluster_kmeans'], cmap=plt.cm.Set1)
axes[0].set_title('Target', fontsize=18)
axes[1].set_title('KMEANS', fontsize=18)
```

```python
neighbors = NearestNeighbors(n_neighbors=8)
neighbors_fit = neighbors.fit(df[FEATURES])
distances, indices = neighbors_fit.kneighbors(df[FEATURES])
```

```python
distances = np.sort(distances, axis=0)
plt.plot(distances[:,1])
plt.show()
```
