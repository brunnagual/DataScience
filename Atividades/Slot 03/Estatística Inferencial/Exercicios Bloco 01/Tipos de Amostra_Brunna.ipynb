Nos passos abaixo ele vaicriar uma função para pegar uma amostra sistemática.

Step serve para mostrar quantos passos ele vai pular

Depois ele criou o indexes que recebe o arrange que vai conter a quantidade que ele vai querer pular de linha dessa amostra (3) passou o lens que é o tamanho do df.

E o step vai receber o step para depois ele passar o tamanho de forma automatica.

Depois ele chama o iloc que é uma localização de indexes e depois ele da retorn na amostra.

```python
# dentro dos index perceba que tem um salto de 3 em 3 de acordo com o parametro.
# A amostra é construido a partir de um salto de 3 em 3 percorrendo todos os dados.
# A amostra sistematica sempre passamos o passo de quanto em quanto que vamos querer.
def amostragem_sistematica(df, step):
  indexes = np.arange(3, len(df), step=step)
  amostra_sistematica = iris_df.iloc[indexes]
  return amostra_sistematica

amostra_sistematica = amostragem_sistematica(iris_df, 3)

display(amostra_sistematica)
```

#Amostragem Clusterizada

Onde nos selecionamos os clusters de uma forma nossa classificamos esses elementos de uma forma aleatoria. Pega a amostra por cluster

```python
#criar uma função que devolva uma amostra clusterizada.
def amostra_cluster(df,n_por_cluster, num_cluster):
  N = len(df)
  K = int(N/n_por_cluster)
  data = None
  for k in range(K):
    sample_k = df.sample(n_por_cluster)
    sample_k['cluster'] = np.repeat(k,len(sample_k))
    df = df.drop(index = sample_k.index)
    data = pd.concat([data, sample_k], axis = 0)

  escolha_cluster_aleatorio = np.random.randint(0,k, size = num_cluster)
  samples = data[data.cluster.isin(escolha_cluster_aleatorio)]
  return(samples)

sample = amostra_cluster(df = iris_df, n_por_cluster=30, num_cluster=2)
sample

```

```python
#Com esse condigo você consegue ver que tem 2 cluster diferentes com 30 informações em cada.
sample.groupby(['cluster'])['cluster'].count()
```
