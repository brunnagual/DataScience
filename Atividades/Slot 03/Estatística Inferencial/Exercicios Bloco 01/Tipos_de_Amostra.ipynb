```python
#criar uma função para pegar uma amostra sistemática

def amostragem_sistematica(df, step):

  indexes = np.arange(3, len(df), step=step)
  amostra_sistematica = iris_df.iloc[indexes]
  return amostra_sistematica

amostra_sistematica = amostragem_sistematica(iris_df, 3)

display(amostra_sistematica)
```

#Amostragem Clusterizada

```python
#criar uma função que devolva uma amostra clusterizada

def amostra_cluster(df, n_por_cluster, num_cluster):
  N = len(df)
  K = int(N/n_por_cluster)
  data = None
  for k in range(K):
    sample_k = df.sample(n_por_cluster)
    sample_k['cluster'] = np.repeat(k, len(sample_k))
    df = df.drop(index = sample_k.index)
    data = pd.concat([data, sample_k], axis = 0)

  escolha_cluster_aleatorio = np.random.randint(0,K, size = num_cluster)
  samples = data[data.cluster.isin(escolha_cluster_aleatorio)]
  return(samples)

sample = amostra_cluster(df = iris_df, n_por_cluster=30, num_cluster=2)
sample
```
