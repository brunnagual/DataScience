```python
import pandas as pd
import numpy as np
import random
import seaborn as sns
```

```python
custo = pd.read_csv('/content/cost-of-living.csv')
custo.head()
```

```python
#distribuio os valores dentro de uma lista.
lista = custo.renda_mensal.values.tolist()
print(lista)
```

```python
#means Ã© uma lista vazia para armazenar as medias.
pop = lista
means = []

for i in range(20):
  sample = random.sample(pop,k = 30)
  means.append(np.mean(sample))
means
#vemos as 20 medias das medias das amostras da renda mensal.
```

```python
np.mean(means)
```

```python
np.mean(lista)
```

```python
sns.histplot(means, kde = True);
```
