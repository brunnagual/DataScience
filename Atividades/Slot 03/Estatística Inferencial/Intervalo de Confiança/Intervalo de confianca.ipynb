```python
import pandas as pd
import numpy as np

from scipy.stats import norm
import scipy.stats as st

import random
```

```python
#
samples = [12,12,13,13,15,16,17,22,23,25,26,27,28,28]
teste_95 = st.t.interval(alpha=0.95, df=len(samples)-1, loc=np.mean(samples), scale=st.sem(samples))

teste_95
```

```python
teste_99 = st.t.interval(alpha=0.99, df=len(samples)-1, loc=np.mean(samples), scale=st.sem(samples))
teste_99
```

```python
np.random.seed(100)
sample_data = np.random.randint(10,50,90)
test_norm_95 = st.norm.interval(alpha=0.95, loc=np.mean(sample_data), scale=st.sem(sample_data))
```

```python
test_norm_95
```

```python
test_norm_99 = st.norm.interval(alpha=0.99, loc=np.mean(sample_data), scale=st.sem(sample_data))
test_norm_99
```
