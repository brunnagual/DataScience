```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("DataPrepFinal.csv")
```

```python
df.head()
# coluna Unnamed: 0 provavelmente seria um id porém o pandas já cria esse campo então pode remover
```

```python
# Aqui ele faz a remoção dos dados que ele não acha relevante. Ex o goal está em varios tipos de moeda e tem a coluna usd_goal_real que está tudo em dolar
# O que torna a analise mais facil pois está comparando tudo da mesma moeda. Mesmo caso no pledged.
df = df.drop(columns=['Unnamed: 0','name','category', 'goal','pledged','usd pledged'])
```

```python
# Nessa função ele pega uma coluna e aplica algo que ele queira.
#Lambda é um indicativo de que eu vou criar minha propria função, não existe uma função especifica. Então vou criar.
# x é cada um dos valores de cada linha. 
#Strip é uma função do pandas para strings onde ela vai remover algo no começo (leading) ou no final (trailing) 
df['usd_goal_real'] = df['usd_goal_real'].apply(lambda x: x.lstrip('USD '))
```

```python
df_invest = pd.read_csv('invested.csv')
```

```python
df_invest.head()
```

```python
df_invest.sort_values(by=['ID'])
```

```python
# Nesse caso os dados estavam dificies de trabalhar tendo casos com o mesmo id mas com estados diferentes.
df_invest.head()
```

```python
#Aqui vai usar o One hot incode vai transformar as colunas em variaveis boleanas.
# Ele pegou as colunas e gerou uma categoria para cada uma delas e transformou o valor em boleano.
df_invest = pd.get_dummies(df_invest,columns=['backedLocation'])
```

```python
df_invest.head()
```

```python
df_invest.sort_values(by=['ID'])
```

```python
#Nesse caso ele deu um group by e depois fez uma agregação com o agg
# No age usou a media no backedlocation usou soma.
df_invest = df_invest.groupby(by=['ID']).agg({'age':'mean','backedLocation_BR':'sum','backedLocation_GBK':'sum','backedLocation_US': 'sum'})
```

```python
df_invest.head()
```

```python
#Aqui ele juntou as colunas todas do lado direito.
df = df.merge(df_invest, how='right', on='ID')
```
