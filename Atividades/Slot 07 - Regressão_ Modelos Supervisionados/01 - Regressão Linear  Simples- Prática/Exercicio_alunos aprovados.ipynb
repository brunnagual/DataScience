#Exercicio

O Ministério da Educação solicitou a você uma previsão do percentual de alunos que serão aprovados no ano de 2017, tendo como base o ano de 2016.

Crie um modelo que consiga prever o valor de aprovação em 2016 com as aprovações de 2015.

Faça todo tratamento de dados aprendido e faça a regressão linear e tire o resíduo para saber em quanto o modelo está errando.


```python
# Importando o arquivo formato xlsx
alunos = pd.read_excel("/content/Copia de aprovacao_alunos .xlsx")
alunos.head()
```

```python
#Criando um gráfico com a reta de regressão
ax = sns.lmplot(x="ano_2015", y="ano_2016", data=alunos)
ax.fig.set_size_inches(12, 6)
ax.fig.suptitle('Reta de Regressão - 2015 x 2016', fontsize=16, y=1.02)
ax.set_xlabels("2015", fontsize=14)
ax.set_ylabels("2016", fontsize=14)
ax;
```

```python
#Tire a correlação entre as variáveis
alunos.corr()
```

```python
#Printe o resultado OLS Regression para entender o R^2, Intercepto e o parâmetro da variável explicativa.
import statsmodels.api as sm
Y = alunos.ano_2016
X = sm.add_constant(alunos.ano_2015)
```

```python
resultado_regressao = sm.OLS(Y,X).fit()
```

```python
print(resultado_regressao.summary())
```

```python
#Criando a predição para a variável Ypredição para a variável Y
alunos['Y_previsto'] = resultado_regressao.predict()
alunos
```

```python
#Criando uma coluna com os resultados do resíduo
alunos["residuo"] = alunos["Y_previsto"] - alunos["ano_2016"]
alunos.head()
```
