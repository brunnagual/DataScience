```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#Importe o arquivo formato xlsx
limite = pd.read_excel("Limite_Credito.xlsx")
limite.head()
```

```python
# Usa a função do pandas para criar dummies e passa como parametro o dataset onde vai pegar todos os dados categoricos
# E transformar em dummies, Prefix para dar o nome de como vai se chamar.
limite = pd.get_dummies(limite, prefix=["Escolaridade"])
limite.head()
```

```python
# Trazer a correlação entre as variaveis
# E podemos ver que Escolaridade_Superior_Pos para o médio está inversamente proporcional.
limite.corr()
```

```python
import statsmodels.formula.api as smf
```

```python
# Trazer todas as variaveis do meu modelo
function = "LimitedoChequeEspecial~Idade+RendimentoTotal+Salario+LimitedeCreditoImediato+Escolaridade_Fundamental_Medio+Escolaridade_Superior_Pos"
```

```python
#modelo MRLS
model = smf.ols(formula=function, data=limite).fit()
print(model.summary())
```

```python
#modelo MRLS
function = "LimitedoChequeEspecial~RendimentoTotal+Salario+LimitedeCreditoImediato+Escolaridade_Fundamental_Medio+Escolaridade_Superior_Pos"
model = smf.ols(formula=function, data=limite).fit()
print(model.summary())
```

```python
#modelo MRLS
# Quando coloco o -1 no final ele não tras o meu intercepto
function = "LimitedoChequeEspecial~RendimentoTotal+Salario+LimitedeCreditoImediato+Escolaridade_Fundamental_Medio+Escolaridade_Superior_Pos -1"
model = smf.ols(formula=function, data=limite).fit()
print(model.summary())
```
