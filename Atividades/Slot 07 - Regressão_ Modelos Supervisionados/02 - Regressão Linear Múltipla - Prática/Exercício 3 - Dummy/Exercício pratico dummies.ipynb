O banco que solicitou o modelo de limite de crédito incluiu novas variáveis no dataset como escolaridade, gênero e Estado do cliente.

Faça as transformações necessárias nessas variáveis Dummies e tire o modelo OLS até que o modelo fique ajustado.


```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#Importe o arquivo formato xlsx
limite = pd.read_excel("Limite_Credito_Dummy.xlsx")
limite.head()
```

```python
limite = pd.get_dummies(data=limite, columns=['Escolaridade', 'Gênero',"Região"], prefix = ['Escolaridade', 'Região',"Gênero"])
limite.head()
```

```python
limite.corr()
```

```python
import statsmodels.formula.api as smf
```

```python
function = "LimitedoChequeEspecial~Idade+RendimentoTotal+Salario+LimitedeCreditoImediato+Escolaridade_Fundamental_Medio+Escolaridade_Superior_Pos+Região_F+Região_M+Gênero_RJ+Gênero_SP"
model = smf.ols(formula=function, data=limite).fit()
print(model.summary())
```
