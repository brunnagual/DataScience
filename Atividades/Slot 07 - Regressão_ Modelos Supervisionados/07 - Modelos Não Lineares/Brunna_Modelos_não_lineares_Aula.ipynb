```python
import pandas as pd
import numpy as np
```

```python
features = pd.read_excel("temps.xlsx")
features.head()
```

```python
features.describe()
```

```python
features.info()
```

```python
# One-hot encode the data using pandas get_dummies
features = pd.get_dummies(features)
features.head()
```

```python
# Labels são os valores que quero predizer
labels = np.array(features["actual"])
# Removendo labels dos features
# axis 1 refere-se a columns
features = features.drop("actual", axis = 1)
# Salvando feature names para usar posteriormente
feature_list = list(features.columns)
# Convertendo numpy para array
features = np.array(features)
```

```python
# Usando Skicit-learn para dar um split nos dados dentro do meu data set de treino.
from sklearn.model_selection import train_test_split
# Fazendo um Split nos dados dentro do training e testing sets.
train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size = 0.25, random_state=42)
```

## Treino com a média

```python
# The baseline predictions are the historical averages
baseline_preds = test_features[:, feature_list.index("average")]
# Baseline errors, and display average baseline error
baseline_errors = abs(baseline_preds - test_labels)
print('Baseline error average: ', round(np.mean(baseline_errors), 2))
```

## Bibiliotecas

```python
from sklearn.datasets import make_moons
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,AdaBoostClassifier,GradientBoostingClassifier, AdaBoostRegressor
from sklearn import metrics
```

## Treino com RandomForestRegressor

```python
# Import o modelo que irei usar
from sklearn.ensemble import RandomForestRegressor
# Instantiate model with 1000 decision trees
rf = RandomForestRegressor(n_estimators=1000, random_state=42)
# Train the model on training data
rf.fit(train_features, train_labels)
```

```python
# Use the forest's predict method on the test data
prediction_rf = rf.predict(test_features)

# Calculate the absolute errors
error_rf = abs(prediction_rf - test_labels)

r_sq = rf.score(features, labels)
# Print out the mean absolute error (mae)
print("R^2:", r_sq)
print("MAE:", metrics.mean_absolute_error(test_labels,prediction_rf))
print("MSE:", metrics.mean_squared_error(test_labels,prediction_rf))
```

## Treino com AdaBoostRegressor

```python
ada = AdaBoostRegressor(n_estimators=100)
ada.fit(train_features, train_labels)
ada_pred = ada.predict(test_features)
```
