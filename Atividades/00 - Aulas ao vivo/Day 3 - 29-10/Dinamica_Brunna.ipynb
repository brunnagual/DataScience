```python
df['Date'] =  pd.to_datetime(df['Date'], format='%d-%m-%Y')
df
```

Plotar um gráfico do tipo scatterplot, em que o eixo x seja referente a meses e o eixo y  a vendas semanais, onde é possível observar as vendas nos dias normais e os feriados

```python
df['month'] = pd.DatetimeIndex(df['Date']).month
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=df, x = "month", y="Weekly_Sales", hue="Holiday_Flag")
```
